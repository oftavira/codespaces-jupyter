```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Create figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate some data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Initialize the plot object
plot = ax.plot_surface(X, Y, Z, cmap='viridis')

# Animation update function
def update(frame):
    # Update the plot data for each frame
    plot.set_array(np.sin(np.sqrt(X**2 + Y**2 + frame/10)))

# Create the animation
animation = FuncAnimation(fig, update, frames=20, interval=1000)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from PIL import Image

# Define the wave parameters
omega = 2 * np.pi  # Angular frequency
k = 0.5 * np.pi    # Wave number
speed = 0.02      # Animation speed

# Create the figure and axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Set up the initial plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(k * X + omega * 0)

surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Animation update function
def update(frame):
    ax.cla()  # Clear the current plot
    Z = np.sin(k * X + omega * frame * speed)
    surf = ax.plot_surface(X, Y, Z, cmap='viridis')
    return surf

# Create the animation
animation = FuncAnimation(fig, update, frames=100, interval=50)

# Save the animation as a video file
animation.save('wave_animatio.mp4', writer='ffmpeg')
```
